## Question 4 ( 2 point)
Design a Python function to check if a number is prime.


```python
# write your answer here
def is_prime(n):

    if int(n) <= 1:
        return False

    if int(n) == 2:
        return True

    if int(n) % 2 == 0:
        return False

    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2

    return True
n = input ('Enter a number')
is_prime(n)
```

## Question 5 ( 2 point)
Write a Python program to concatenate following dictionaries to create a new one.

Sample Dictionary : dic1={1:10, 2:20}, dic2={3:30, 4:40}, dic3={5:50,6:60},

Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}


```python
# write your answer here

dic1 = {1:10, 2:20}
dic2={3:30, 4:40}
dic3={5:50,6:60}

combined_dict_update = {}
combined_dict_update.update(dic1)
combined_dict_update.update(dic2)
combined_dict_update.update(dic3)

print("\nResult using update() method:")
print(combined_dict_update)
```
